```python
import matplotlib.pyplot as plt
import numpy as np

import torchvision.transforms as T

import _frameload
```

```python
images_train, labels_train, images_test, labels_test = _frameload.main('train/', n_train=30000, n_test=10000, processes=6)
```

```python
transform = T.Compose([
    T.functional.autocontrast,
    T.ToPILImage()
])

def showRandon(images, labels, n_subplots=8):
    fig, axes = plt.subplots(1, n_subplots, figsize=(n_subplots * 2, 2))
    idximg = np.random.randint(0, len(images), n_subplots)
    for i, ax in enumerate(axes):
        ax.imshow(transform(images[idximg[i]]))
        ax.set_title('Anime' if labels[idximg[i]] == 1. else 'Photo')
    
    plt.tight_layout()
    plt.show()
```

```python
for i in range(4):
    showRandon(images_train, labels_train)
```

```python
for i in range(4):
    showRandon(images_test, labels_test)
```
